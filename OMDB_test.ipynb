```python
# Dependencies
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import requests

# API key
from config import api_key
import json
```

```python
# go to OMDB and get data 

movies_2018 = []

movies2018 = pd.read_csv("2018_movie.csv")

for movie in movies2018["Title"]:
    try:
        url = f"http://www.omdbapi.com/?t={movie}&apikey={api_key}"
        response = requests.get(url).json()

        title = response["Title"]
        year = response["Year"]
        released = response["Released"]
        runtime = response["Runtime"]
        genre = response["Genre"]
        awards = response["Awards"]
        metascore = response["Metascore"]
        imdbRating = response["imdbRating"]
        imdbVotes = response["imdbVotes"]
        imdbID = response["imdbID"]
#         budget = response["Budget"]
#         boxoff = response["BoxOffice"]
        
        movies_2018.append({"Title": title,
                            "Year": year,
                            "Released Date": released,
                            "Runtime": runtime,
                            "Genre": genre,
                            "Awards": awards,
                            "MetaScore": metascore,
                            "imdbRating": imdbRating,
                            "imdbVotes": imdbVotes,
                            "imdbID": imdbID})
#                             "Budget": budget,
#                             "Box Office": boxoff})
        
        print(f"fetching data for {movie}")
                            
    except:
        print("movie not found")
        pass

movies_2018_df = pd.DataFrame.from_dict(movies_2018)
movies_2018_df.head()
```
